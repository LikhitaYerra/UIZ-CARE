Import of the libraries

```python
import pandas as pd
from difflib import SequenceMatcher 
```

Loading of the dataset using the databases of WHO

```python
# Load the Excel file into a pandas DataFrame
df = pd.read_excel(r"C:\Users\lecka\Downloads\Clinic of IA\simpletabulation.xlsx", usecols=['Title', 'Code'])

#Data cleaning
df=df.dropna()
df=df.drop_duplicates()

# Remove "-" characters from the 'Title' column
df['Title'] = df['Title'].str.replace('-', '').str.strip()
```

```python
# Create the data dictionary DataFrame with only 'Title' and 'Code'
list_code=list(df['Code'])
list_title=list(df['Title'])

dict_data = {k: v for k, v in zip(list_code, list_title)}
dict_data
```

Creation of the classe for the recommendation system

```python
class SystemeRecommandation: 

    def __init__(self, data): 

        self.data = data 

  

    def recommander(self, input_description, threshold=0.6): 

        suggestions = [] 

  

        for code, description in self.data.items(): 

            similarity_ratio = SequenceMatcher(None, input_description, description).ratio() 

             

            if similarity_ratio >= threshold: 

                suggestions.append((code, description, similarity_ratio)) 

  

        suggestions.sort(key=lambda x: x[2], reverse=True)  # Tri par similarit dcroissante 

        return suggestions 
```

```python
def applicate_recommandation_systeme(data):
 systeme_recommandation = SystemeRecommandation(data) 
    # Exemple de recommandation pour une description donne 

 input_description = input("Enter the title of the diseases")
 recommandations = systeme_recommandation.recommander(input_description) 

 if recommandations: 

        print(f"Recommandations pour '{input_description}':") 

        for code, description, similarity in recommandations: 

            print(f"Code: {code}, Title: {description}, Similarit: {similarity:.2f}") 

 else: 

        print("Aucune recommandation trouve.") 
```

```python
applicate_recommandation_systeme(dict_data)
```
